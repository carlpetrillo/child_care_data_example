This is a quick and dirty example of how one might visualize supply and demand for child care in Seattle.

```python
import numpy as np
import pandas as pd
import requests as r
```

```python
acs_url = 'https://api.census.gov/data/2019/acs/acs5?get=NAME,B01001_001E&for=tract:*&in=state:53%20county:033'
```

```python
acs_response = r.get(acs_url)
```

```python
acs = pd.DataFrame(acs_response.json())
acs.columns = acs.iloc[0]
acs = acs.drop(0).reset_index().drop('index', axis=1)
acs
```
